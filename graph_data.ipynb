```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
right = 'all dfs/for X/Айкун_правильно.txt'
df_right = pd.read_csv(right, delimiter=',')
ind_right=range(11331,14925)

wrong = 'all dfs/for X/Айкун_неправильно.txt'
df_wrong = pd.read_csv(wrong, delimiter=',')
ind_wrong=range(6600,9000)
```

```python
plt.figure()
plt.figure(figsize=(11,7))
plt.plot(df_right['ay(g)'][ind_right], label='AccY')
plt.plot(df_right['az(g)'][ind_right], label='AccZ')
plt.plot(df_right['ax(g)'][ind_right], label='AccX')
plt.xlabel('row number', fontsize=20)
plt.ylabel('Acceleration XYZ, 9.8 m/s²', fontsize=20)
plt.title(r'Один эпизод правильно выполненного упражнения', fontsize=20)
plt.legend()
```

```python
plt.figure()
plt.figure(figsize=(11,7))
plt.plot(df_wrong['ay(g)'][ind_wrong], label='AccY')
plt.plot(df_wrong['az(g)'][ind_wrong], label='AccZ')
plt.plot(df_wrong['ax(g)'][ind_wrong], label='AccX')
plt.xlabel('row number', fontsize=20)
plt.ylabel('Acceleration XYZ, 9.8 m/s²', fontsize=20)
plt.title(r'Один эпизод неправильно выполненного упражнения', fontsize=20)
plt.legend()
```

```python
def to_plot_acc(file, ind, title):
    df = pd.read_csv(file, delimiter=',')
    plt.figure()
    plt.figure(figsize=(11,7))
    plt.plot(df['ay(g)'][ind], label='AccY')
    plt.plot(df['az(g)'][ind], label='AccZ')
    plt.plot(df['ax(g)'][ind], label='AccX')
    plt.xlabel('row number', fontsize=20)
    plt.ylabel('Acceleration XYZ, 9.8 m/s²', fontsize=20)
    plt.title(title, fontsize=20)
    plt.legend()
```

```python
to_plot_acc('all dfs/for X/Айкун_правильно.txt', range(0,10000), 'Несколько эпизодов правильно выполненного упражнения')
to_plot_acc('all dfs/for X/Айкун_неправильно.txt', range(0,10000), 'Несколько эпизодов неправильно выполненного упражнения')
```
